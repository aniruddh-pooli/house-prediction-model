```python
def convertRange(x):
    temp = x.split('-')
    if len(temp)==2:
        return (float(temp[0])+float(temp[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
df['total_sqft'] = df['total_sqft'].apply(convertRange)
```

```python
df['ppsm'] = df['price'] * 100000 / df['total_sqft']
```

```python
def remove_outlier_sqft(df):
    df_output = pd.DataFrame()
    for key,subdf in df.groupby('location'):
        m = np.mean(subdf.ppsm)
        st=np.std(subdf.ppsm)

        gen_df = subdf[(subdf.ppsm>(m-st))&(subdf.ppsm<=(m+st))]
        df_output = pd.concat([df_output,gen_df],ignore_index=True)
    return df_output
df = remove_outlier_sqft(df)
df.describe()
```

```python
def bhk_remove_outlier(df):
    exclude_indices = np.array([])
    bhk_stats = {} 
    for location, location_df in df.groupby('location'):
        for bhk, bhk_df in location_df.groupby('bhk'):
            if bhk not in bhk_stats:
                bhk_stats[bhk] = {
                    'mean': np.mean(bhk_df['ppsm']),
                    'std': np.std(bhk_df['ppsm']),
                    'count': bhk_df.shape[0]
                }
        
        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk - 1)
            if stats and stats['count'] > 5:
                exclude_indices = np.append(
                    exclude_indices,
                    bhk_df[bhk_df['ppsm'] < stats['mean']].index.values
                )
    
    return df.drop(exclude_indices, axis='index')
```

```python
df = bhk_remove_outlier(df)
```

```python
x = df.drop(columns=['price'])
y = df.drop(columns=['location','total_sqft','bath','bhk'])
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.preprocessing import OneHotEncoder 
from sklearn.preprocessing import StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=2,random_state=0)
```

```python
column_trans = make_column_transformer((OneHotEncoder(sparse_output=False),['location']),remainder='passthrough')
```

```python
scaler = StandardScaler()
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans,scaler,lr)
```

```python
print(x_train.columns)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred_lr=pipe.predict(x_test)
```

```python
r2_score(y_test,y_pred_lr) 
```

```python
lasso= Lasso()
```

```python
pipe = make_pipeline(column_trans,scaler,lasso)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred_lasso = pipe.predict(x_test)
r2_score(y_test,y_pred_lasso)
```

```python
ridge = Ridge()
pipe = make_pipeline(column_trans,scaler,ridge)
pipe.fit(x_train,y_train)
y_pred_ridge = pipe.predict(x_test)
r2_score(y_test,y_pred_ridge)
```

```python
print("No Regularization:" ,r2_score(y_test,y_pred_lr))
print("lasso:",r2_score(y_test,y_pred_lasso))
print("ridge:",r2_score(y_test,y_pred_ridge))

```
